### One sample t-test:

- The one sample t-test is a statistical test used to determine whether the
mean of a single sample differs significantly from a known or hypothesized
population mean.


- It's typically used when you have a single group of
subjects and you want to compare their mean score on a continuous
variable to a known or hypothesized population mean.



###  Two independent sample t-test:

- The two independent sample t-test is used to determine whether the
means of two independent groups are significantly different from each
other. 


- It's commonly used in experimental research where there are two
separate groups (e.g., experimental group and control group), and you
want to compare their means on a continuous outcome variable.


###  ANOVA (Analysis of Variance):

- ANOVA is a statistical method used to compare the means of three or
more independent groups to determine if there are statistically significant
differences between them.


- It extends the two sample t-test to multiple
groups. ANOVA tests the null hypothesis that all group means are equal
against the alternative hypothesis that at least one group mean is different.

**One sample t-test example**

- Odometers are utilized to gauge automobile mileage,
but how accurately do they reflect the actual distance
traveled? Let's consider a scenario where 12 cars
travel a predetermined distance of exactly 10 miles,
and the recorded mileage figures from their
odometers are as follows

: 9.8, 10.1, 10.3, 10.2, 9.9,10.4, 10.0, 9.9, 10.3, 10.0, 10.1, 10.2

- Using a significance level of 0.01, we aim to
determine whether the odometers can be trusted.
Before proceeding with the analysis, it's essential to
validate the assumptions underlying the statistical
test.

### a) Run a hypothesis testing to see if the odometer is not accurate in terms of the population mean?

$H_0$ :$\mu = 10$ 

$H_1$ :$\mu \ne 10$ 

```r
#one sample t-test for (a)

mile= c(9.8, 10.1, 10.3, 10.2, 9.9,10.4, 10.0, 9.9, 10.3, 10.0, 10.1, 10.2)
t.test(mile, alternative="two.sided", mu=10)

# t statistic
#1.86 
```

### b) Run a hypothesis testing to see if the odometer is overly gauge the actual mileage? 

$H_0$ :$\mu = 10$ 

$H_1$ :$\mu > 10$ 

```r
#one sample t-test for (b)

t.test(mile, alternative="greater", mu=10)
```

# Simultaneous tests on several variables (Multivariate tests) 


 When data are collected for multiple variables on the same
sample units, it is common practice to analyze each variable
individually for significance tests. For instance, in a scenario
with sample units divided into two groups, the difference in
means for each group can be assessed separately for each
variable. 

However, this straightforward approach has a
drawback due to the repeated use of significance tests, each
with its own probability of yielding an incorrect conclusion. As
elaborated further (refer to Section 4.4 of the book), the
likelihood of erroneously detecting at least one significant
difference increases with the number of tests conducted,
potentially reaching unacceptable levels.

In addtion, the variables may be related, and multiple individual tests do not involve any correlation in the calculation. ..

```r
#Two sample hotelling t test
#grp takes only two values 1, 2

index.s=sparrow_dat$Survivorship == 'S'
sparrow_dat$grp = rep(1, length(index.s))
sparrow_dat$grp[!index.s] = 2

#sparrow_dat

#Total_length, Alar_extent, L_beak_head :
#Function for two sample Hotelling T test in package "MVtests" or ...
TwoSamplesHT2(sparrow_dat[,c(2,3,4)], group=sparrow_dat$grp)

#Test statistic T2 = 1.4638
#F-value = 0.467
#p-value = 0.706624
#conclusion: fail to reject H0. There is no sufficient evidence that the popu
# means of Total_length, Alar_extent, L_beak_head are different between survive
```

```r
#Self defined Function for two sample Hotelling t test
TwoSampleT2Test = function(X, Y)
{
nx = nrow(X)
ny = nrow(Y)
#The difference between two sample mean vectors
delta = colMeans(X) - colMeans(Y)
#This is the number of variables
p = ncol(X)
Sx = cov(X)
Sy = cov(Y)
#To obtain the pooled covaraince matrix of two groups
S_pooled = ((nx-1)*Sx + (ny-1)*Sy)/(nx+ny-2)
#Hotelling T
T_squared = (nx*ny)/(nx+ny) * t(delta) %*% solve(S_pooled) %*% (delta)
#F-statistic
statistic = T_squared * (nx+ny-p-1)/(p*(nx+ny-2))
p_value = pf(statistic, p, nx+ny-p-1, lower.tail=FALSE)
return(list(Test.Statistic=statistic, p.value=p_value, df1=p, df2=nx+ny-p-1))
}
#How to use it: to compare the population mean vector of X group vs Y group
#TwoSampleT2Test(X, Y)
#Data for the the variables "Total_length, Alar_extent, L_beak_head " of survi
S.dat = sparrow_dat[index.s,c(2,3,4)]
#Data for the the variables "Total_length, Alar_extent, L_beak_head " of non-s
NS.dat = sparrow_dat[!index.s,c(2,3,4)]
TwoSampleT2Test(X=S.dat, Y=NS.dat)
```

```r
#MANOVA
#Test if the population mean vector of Al, Mg, Ca, Na, Mn are the same or not
fit = manova(cbind(Al, Mg, Ca, Na, Mn) ~ Site , data = Pottery2 )
summary(fit)


#?manova
#Test statistic = Pillai = 1.89
#P-value = 5.066e-15 (Reject H0)
#Conclusion: Reject H0. There is sufficient evidence in the data to support H1
#that the population mean vector of Al, Mg, Ca, Na, Mn depends on the site.
#Normality check


library(psych)
mardia(fit$residuals)
#The QQ normal plot for the residual vector indicates the normality assumpotio
#Or ... the error terms of the data are from a normal distribution approximate
##Or ... the error terms of the data are slightly non- normaldistribution.
#Check the assumption of the homogeneity of the variance-covariance matrices
#Boxs M-test


library(MVTests)
#function "with" : to associate a data set with a R function
print("BoxM test result:")
with(Pottery2, BoxM(cbind(Al, Mg, Ca, Na, Mn) , group = Site ))


# For BoxM test to check the equality of population covariance matrix of different group,
# Test statistic =\infinity (some issue usually it is a valid number)
# p-value = 0.0 
# Conclusion : reject H0, There is sufficient evidence that population covariance 
# matrices of different group are not the same.
```
