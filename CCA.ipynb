# Canonical Correlation Analysis (CCA)

example data

- group 1 ( sales performance ) : growth, profit , new
- group 2 (intelligence) : create, mech, abs, math 

**purpose : to see how sale performance and intelligence are related.**

```r
data = read.csv("sales.csv")
str(data)
```

```r
library(CCA)
```

```r
# CCA analysis

# sales performance 
X=data[,c(1:3)]

# intelligence 
Y=data[,c(4:7)]

cor(X,Y)

M=matcor(X,Y) #not very useful in this case, not necerray to use it
M

matplot(X,Y)
```

```r
u1 = (cca.result$scores$xscores)[,1]
v1 = (cca.result$scores$yscores)[,1]

cor(u1,v1)
```

```r
plt.cc(cca.result)
```

```r
# to determine the number of canonical pairs 
library(CCP)

#the number of observations
# n = length(X[,1])
n=nrow(X)

#calculate p-values using F-approximation
p = ncol(X) # 3 # number of x variables ("performance")
q = ncol(Y) # 4 # number of y variables ("intelligence")

rho = cca.result$cor 
p.asym(rho,n,p,q,tstat="Wilks")

#the p-value of the first set canonical variate is 0.0000

#because p-value of the last set is 2.7835363-02 , all tjree set of canonical variates are related./

# for example, if the p-value for the first(1 to 4:) is 0.0002 and the second p-value (0.394)
# then , there is sufficient evidence that X variables and Y variables are related.
# but the second to the last sets of canonial variates are not significant.
```

```r
# wine data 

loc = "http://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
data = read.table(loc,sep=",")
str(data)

# the first column is the class label, the rest are the features
dat= data[,c(2:14)]
```

```r
X=dat[,c(1:5)]
Y=dat[,c(6:13)]
cca.1 = cc(X,Y)

cca.1$cor
```

```r
n=nrow(X)

p=ncol(X)
q=ncol(Y)

rho = cca.1$cor
p.asym(rho,n,p,q,tstat="Wilks")

#only three paris of canonical variates are significant, because the p-value of "4 to 5" : 0.19 
```
