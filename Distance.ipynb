# Mahalanobis Distance

$$d(x,y)=\sqrt{(x-y)^T\Sigma^{-1}(x-y)}$$
where $\Sigma$ stands for the covariance. 

Example : lex X=(4,3) and Y=(1,7). 
The covariance matrix of the underlying bivariate distribution is diagnal matrix(4,1)

Find the Mahalanobis Distance

```r
#Find the mahalanobis distance
sum(abs(p-q))
```

```r
x=c(4,3)
y=c(1,7)

S=matrix(c(4,0,0,1),nrow=2)

#Diagonal matrix
S=diag(c(4,1))
S

#Find the Mahalanobis distance bewteen x,y given the covariance matrix S
A=sqrt(t(x-y)%*%solve(S)%*%(x-y))
A
```

Example : Let X=(4,3) and Y=(1,7). the covariance matrix of the underlying bivariate distributoin is diagnal matrix(32,8). what is the Mahalanobis Distance between x and y?

```r
x=c(4,3)
y=c(1,7)

S=matrix(c(4,0,0,1),nrow=2)

#Diagonal matrix
S=diag(c(32,8))
S

#Find the Mahalanobis distance bewteen x,y given the covariance matrix S
sqrt(t(x-y)%*%solve(S)%*%(x-y)) # 1.510381


1.510381*sqrt(8) # it is the same result as A (previous question)
```

```r
# example find the squared Mahalanobis distance matrix of all observations of survived cases in the sparrow data.
# let the sample covariance be the estimate of the unknown covariance matrix.
data=read.csv("Bumpus_sparrows.csv")
head(data)

#Data for sparrow of survived cases, without the categorical variable
sdat=data[data$Survivorship=="S",-1] #remove the first column
sdat


#find the dist matrix for the squared mahalanobis distance
#Margin =1 : line by line, 2= column by column
#returns the squared mahalanobis distance matrix by mahalanobis distance in R
covMat = cov(sdat)
m.dist.mat=apply(sdat,MARGIN=1,function(i){mahalanobis(sdat,i,cov=covMat)})

#round(m.dist.mat,1)

#visualize it 
heatmap(m.dist.mat,Colv = NA,Rowv = NA,scale='none')
```

# The mantel randomization test 

```r
library(ade4)

ozone = read.csv("ozone.csv",header = TRUE)
 

#the purpose is to see if the distance in the pollution level and the geographical distance are related or not.
#we can use mantel test 

#distance objects returned by dist function
g.dist = dist(cbind(ozone$Lon, ozone$Lat))
ozone.dist = dist(ozone$Av8top)

set.seed(217)
mantel.rtest(g.dist,ozone.dist, nrepet = 9000)
```

```r
cor(g.dist, ozone.dist)
#simulated p-value : 0.03199644
#reject H0(nor related). there is sufficient evidence in the data to 
#support H1 that the distance in ozone is related to the geographical distance.

#The correlation between two dist object is 0.1636308 
```

# Multivariate outliers

1. determine any outliers using all quantative variables of the bird data.

2. using the sample covariance of all quantiataive variables, 

3. for survival 


```r
# load the data

sparrow.dat=read.csv("Bumpus_sparrows.csv")

 
dat= sparrow.dat[,-1]
head(dat)

#step 1 : calculate the manhanobis distance to the center of the data

mean.vec = colMeans(dat)
mean.vec

cov.dat=cov(dat)
```

```r
#squared mahalanobis distance
dist.vec=mahalanobis(dat,center=mean.vec,cov=cov(dat))

hist(dist.vec)
# if the data are from a multivariate normal , the distribution of squared mahalanobis distance has a Chisq distribution 
# with the degrees of freedom = the number of variables. 

# if the level of significance = 0.01 , if p-value of an observation is below the level of significance, then it is identified as an outlier based on this method.
# then it is identified as an outlier based on this method ..
```

```r
#P (chisq distribution > the observed distance)
pval.vect = 1-pchisq(dist.vec, df=5)

round(pval.vect ,5)

# the number of outliers at the significance level of 0.01
sum(pval.vect < 0.01)

#see which is (are) outliers
dat[pval.vect < 0.01,]
```
