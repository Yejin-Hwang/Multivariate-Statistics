# Linear transformation

A linear transformation from x to y, where $a$ and $b$ are constants:
$y= a+bx$

Example : temperature data(data in Celsius to data in Fahrenheit)
    $$
    data in Fahrenheit = 9/5 * data in Celsius + 32
    $$
    
    Suppose that we know that the mean temperature of 20 data is 10 Celsius and the standard deviation is 5 Celsius. Find the mean temperature in Fahrenheit; find the standard deviation for the temperature data in Fahrenheit.
    
    The mean = apply the linear transformation to the mean = 9/5 * 10 + 32 = 50 Fahrenheit
    
    The standard deviation of the data in Faherenheit = 9/5 * 5 = 9

A correlation matrix is a matrix showing correlation coefficients between variables.

#A correlation matrix is a matrix showing correlation coefficients between variables.

Comments about the correlation:

#Comments about the correlation coeffieicents

Correlation r is always -1 < r < 1. The extreme value r = -1 or r = 1 occur 
when the points in the scatterplot lie exactly along a straight line. 

Linear Correlation Coefficient r measures "only the linear relationship" between two quantitative variables. 
Correlation does "not" describe "curved relationships" between variables, no matter how strong they are.


If there is a strong positive linear relationship between two variables, the value of r is close to +1. If there is a strong negative linear relationship between two variables, the value of correlation r is close to -1. When two variables have weak or no association, the value of r is close to 0.


Like the mean and standard deviation, the Linear Correlation Coefficient is sensitive to outliers and extreme values.

Note: the Linear Correlation Coefficient r only measures the strength of a linear

# Graphical tools for multivariate data

1. Chernoff faces 

Chernoff faces is to present multivariate variables simultaneously by mapping elements of the human face, such as ears, hair, eyes, and nose, according to numerical values in a dataset. The underlying idea is that since we are adept at interpreting facial expressions in everyday life, we should likewise be capable of discerning subtle variations when they correspond to data points.

* Package aplpack

2. Star plots

Star plots (Chambers, et al., 1983, pp. 158-162) are a useful way to display multivariate observations with an arbitrary number of variables. Each observation is represented as a star-shaped figure with one ray for each variable. For a given observation, the length of each ray is made proportional to the size of that variable. Star plots differ from glyph plots in that all variables are used to construct the plotted star figure; there is no separation into foreground and background variables.

3. Profile plots or Parallel Coordinate Plots

# Profile plots or Parallel Coordinate Plots

A parallel coordinate plot illustrates each row in a dataset as a line or profile, with each feature of a row represented by a point on the line. Although parallel coordinate plots may resemble line charts, their approach to translating data into a plot differs significantly. This visualization technique is well-suited for representing multivariate, quantitative data, making it ideal for comparing numerous variables and discerning the relationships among them.

# Multivariate skewness and kurtosis 

The univariate version: 

The variance of a random variable:

$$
E(X-EX)^2
$$


Skewness is related to 3rd momement of a random variable
population skewness is related to $$EX^3$$

The sample skewness: (공식)



kurtosis is related to 4th moment of a random variable

The sample kurtosis (different version exist):

The skewness is around zero, if a distribution is symmetric about the mean. a positive
skewness indicates that the distribution is skewed to the right.
A negative skewness indicates that the distribution is skewed to the left.
Kurtosis measures how peaked the distribution is. The kurtosis of a normal distribution is 0.
Kurtosis characterizes the relative peakedness or flatness of a distribution compared to the
normal distribution. If kurtosis is above zero, the tail of the distribution is longer than a normal
distribution. If kurtosis is belwo zero, the tail of the distribution is shorter than a normal
distribution


The multivariate version: 

Population Multivariate skewness and Kurotosis: 


The sample skewness and kurtosis: 
for a sample of size n,


In different statistical packages, they may use different formula for calcualtion. The results may
be slightly different. Multivariate skewness and kurtosis measure the same shape
characteristics as in the univariate case. However, instead of making the comparison of the
distribution of one variable against a univariate normal distribution, they are comparing the joint
distribution of several variables against a multivariate normal distribution.



# Calculate the univarate and multivariate skewness and kurtosis 

```r
#Calculate the univarate and multivariate skewness and kurtosis 

#For mardia measure
#package psych

#skew(x, na.rm = TRUE, type=3)
#kurtosis (x, na.rm = TRUE, type=3)
#trardia(x,na.rm = TRUE,pLot=TRUE)

library(psych)

#Example 1: find the skewness of "Total_Length"

head(dat)
skew(dat$Total_length, na.rm=TRUE, type=1)
skew(dat$Total_length, na.rm=TRUE, type=2)
skew(dat$Total_length, na.rm=TRUE, type=3)

# The skewness of "Total_length" is 0.143.

#To calculate the skewness of all quantitative variables
apply(dat[,-1],2, skew)
skew(dat[,-1])

#the univariate skewness for each variable............ ( not multivariate !! **)
```
